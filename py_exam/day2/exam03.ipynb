```python
# 문제 : 1부터 5까지의 합을 출력해주세요.(개념 : while)
i = 1
sum = 0
while i <= 5:
    sum += i
    i += 1
print(sum)
```

```python
# 문제 : while문으로 구구단 1단 ~ 9단 출력

dan = 1
while dan <= 9:
    i = 1
    while i <= 9:
        print("{} * {} = {}".format(dan, i, dan * i))
        i += 1
    dan += 1
```
