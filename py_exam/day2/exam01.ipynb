```python
# 문제 - 사용자에게 문장 1개를 입력받아서, ','를 기준으로 나눠주세요. split() 함수

line = input("문장을 입력 해주세요")#.split(',')
line = line.split(',')             # 리스트로 반환
print(line)
print(line[0])
print(line[1])
print(line[2])
print(line[3])
print(line[4])
```

```python
# 문제 - 사용자에게 문장 1개를 입력받아서, strip 한 결과를, 다시 ','를 기준으로 나눠주세요.

line = input("문장을 입력 해주세요")
line = line.strip()     # 문장 쪼개기 전에
line = line.split(',')
```

```python
line = input("문장을 입력해주세요").strip().split(',')
print(line)
print(line[0])
print(line[1])
print(line[2])
```

```python
# 문제 - 사용자에게 숫자 2개를 입력받아서, 더한 결과를 출력해주세요.

line = input("숫자를 입력해주세요").strip().split(',')
print(line[0] + line[1])
# line[0] = int(line[0])
# line[1] = int(line[1])
print(int(line[0]) + int(line[1]))
```

```python
# 문제 : 사용자에게 숫자 3개를 입력받아서, 더한 결과를 출력해주세요. map, strip를 사용해주세요.

line = input("숫자를 입력해주세요").strip().split(',')
# line[0] = int(line[0])
# line[1] = int(line[1])
# line[2] = int(line[2])
line = list(map(int, line))
print(line[0] + line[1] + line[2])

# a, b, c = map(int, input("숫자를 입력해주세요").strip().split(','))
# print(a + b + c)

# map 함수는 리스트의 모든 요소에 함수를 적용할 때 사용.
```
