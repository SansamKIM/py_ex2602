```python
# 정규표현식(정규식)

text = "안녕하세요. 저는 15살입니다."

print("15" in text)
```

```python
# 정규식에 관한 라이브러리
import re

# \d는 숫자
# +는 1개 이상

text = "안녕하세요. 저는 15살입니다. 저는 2009년생입니다."

pattern = r"\d+"    # r은 raw string \(특수문자)★
result = re.findall(pattern, text)
print(result)
```

```python
sample = "abc123def456"

print(re.findall(r"\d+", sample))

print(re.findall(r"[a-z]+", sample))
```

```python
# 특정 형식 찾기

data = "test123@gmail.com"

email_pattern = r"\w+@\w+.\w+"
# \w : 문자 또는 숫자
# + : 1개 이상
# @ : 말 그대로 @
# \. : 점(특수문자이기 때문에 \를 붙인다)

emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환

test = "가격은 1200원 입니다."

new_text = re.sub(r"\d+", "", test)
print(new_text)
```

```python
# 정규식 vs 일반 문자열 처리

test = "몸무게: 80kg, 키: 180cm"

test.split(',')
numbers = re.findall(r"\d+", test)
print(numbers)
```
