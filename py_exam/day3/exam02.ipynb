```python
# 문제 : 함수를 사용해서 코드량을 확 줄여주세요.

# 자주 사용하는 코드를 함수로 만들어두면 참 편하다.
# 이름이 `print8Dan`인 함수를 정의한다. 또는 만든다.

def print8dan():
    dan = 8
    i = 1
    while i <= 9:
        print("{} * {} = {}".format(dan , i, dan * i))
        i += 1

print("== 1번째 99단 8단 출력 ==")
print8dan()
```

```python
# 문제 : 함수를 사용해서 코드량을 확 줄여주세요.

# 자주 사용하는 코드를 함수로 만들어두면 참 편하다.
# 이름이 `print8Dan`인 함수를 정의한다. 또는 만든다.
# 참고로 `(dan)` 안에 있"는 `dan` 변수는 함수 외부의 `dan` 변수와 아무런 관련이 없다.

def print99dan(dan):
    i = 1
    while i <= 9:
        print("{} * {} = {}".format(dan, i, dan * i))
        i += 1

print99dan(1)
print99dan(2)
print99dan(3)
print99dan(4)
print99dan(5)
print99dan(6)
print99dan(7)
print99dan(8)
print99dan(9)
```

```python
# 문제 : 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.

def is_prime_number(num):
    if num == 1:
        return False
    
    i = 2
    while i < num:
        if num % i == 0:
            return False
        i += 1

    return True

print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6는 소수입니다 : {}".format(is_prime_number(6)))
print("7는 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```

```python
# 문제 : 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요. ver2

def get_divisors_sum(num):
    s = 0
    i = 1
    while i <= num:
        if num % i == 0:
            s += i
        i += 1
    return s

def is_prime_number_v2(num):
  return get_divisors_sum(num) == num + 1

a = 46
is_prime_number_v2(a)
```

```python
# 문제 : for문으로 1부터 n사이에 존재하는 소수의 합을 반환하는 함수 구현

def is_prime_number(number):
  if number <= 1:
    return False

  for i in range(2, number):
    if number % i == 0:
      return False
  
  return True

# 1부터 n까지 사이의 모든 소수의 합을 반환하는 함수 구현
def get_1_to_n_prime_sum(n):
    s = 0

    for i in range(1, n + 1):
        if is_prime_number(i):
            s += i
    return s

'''
def get_1_to_n_prime_sum(n):
    s = 0

    for number in range(1, n + 1):
        if number <= 1:
            continue

        is_prime = True
        for i in range(2, number):
            if number % i == 0:
                is_prime = False
                break

        if is_prime:
            s += number

    return s
'''

# 함수 사용
n = 5
print("1부터 {}사이에 존재하는 모든 소수 의 합 : {}".format(n, get_1_to_n_prime_sum(n)))

n = 100
print("1부터 {}사이에 존재하는 모든 소수 의 합 : {}".format(n, get_1_to_n_prime_sum(n)))

n = 1000
print("1부터 {}사이에 존재하는 모든 소수 의 합 : {}".format(n, get_1_to_n_prime_sum(n)))
```
