```python
# 클래스(class) : 객체(instance)를 표현하기 위한 문법의 일종, 객체를 만들기 위한 설계도

# 전사, 마법사, 자동차, 비행기 -> 객체(object) = 현실, 개념, 가상, 개념 가능

# 객체(instance)는
# 속성과(attribute) = 머리 색, 장비, 갑옷, ...           - 명사적
# 메서드(method) = 공격, 방어, 이동, ...                 - 동사적

class Person:
    def hello(self):
        print('hi')

jin = Person() # jin은 Person 클래스(class)의 객체(instance)이다.


```

```python
class Person:
    def hello(self):
        print('hello')
    
    def hi(self):
        self.hello()
jin = Person()
jin.hi()
# self : class 내의 일종의 리모컨
```

```python
# 객체(instance) 속성(attribute) 사용 방법
class Person:
    def __init__(self):
        self.hello = '안녕하세요' # 인스턴스 속성

    def hi(self):
        print(self.hello)

jin = Person() # 이 때 __init__() 메서드는 이미 실행된다.
jin.hi()
```

```python
class Person:
    def __init__(name, age, hobby):
        self.name = name
        self.age = age
        self.hobby = hobby
    
    def introduce(self):
        print('안녕하세요, 저는{}입니다. 나이는 {}살이고, 취미는 {}입니다.')

jin = Person('진', 20, '코딩')
jin.introduce()
```
