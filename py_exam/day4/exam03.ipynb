```python
# 클래스(class) : 객체(instance)를 표현하기 위한 문법의 일종, 객체를 만들기 위한 설계도

# 전사, 마법사, 자동차, 비행기 -> 객체(object) = 현실, 개념, 가상, 개념 가능

# 객체(instance)는
# 속성과(attribute) = 머리 색, 장비, 갑옷, ...           - 명사적
# 메서드(method) = 공격, 방어, 이동, ...                 - 동사적

class Person:
    def hello(self):
        print('hi')

jin = Person() # jin은 Person 클래스(class)의 객체(instance)이다.


```

```python
class Person:
    def hello(self):
        print('hello')
    
    def hi(self):
        self.hello()
jin = Person()
jin.hi()
# self : class 내의 일종의 리모컨
```

```python
# 객체(instance) 속성(attribute) 사용 방법
class Person:
    def __init__(self):
        self.hello = '안녕하세요' # 인스턴스 속성

    def hi(self):
        print(self.hello)

jin = Person() # 이 때 __init__() 메서드는 이미 실행된다.
jin.hi()
```

```python
class Person():
  def __init__(self, name, age, hobby): # 이 안에 있는 속성은 인스턴스 속성이다.
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.hobby = hobby

  def introduce(self):
    print("{}, 저는 {} 입니다.".format(self.hello,self.name)) # 클래스 내부에서 접근

jin = Person('진', 22, '운동')
jin.introduce()

paul = Person('폴', 33, '요리')

print("== 진 ==")
print('이름 : {}'.format(jin.name)) # 클래스 외부에서 접근
print('나이 : {}'.format(jin.age))
print('취미 : {}'.format(jin.hobby))

print("== 폴 ==")
print('이름 : {}'.format(paul.name))
print('나이 : {}'.format(paul.age))
print('취미 : {}'.format(paul.hobby))
```

```python
# 클래스 외부에서는 접근 불가, 클래스 내부에서만 사용 가능한 private attribute

class Person():
  def __init__(self, name, age, hobby, weight):
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.hobby = hobby
    self.__weight = weight

jin = Person('진', 22, '운동', 78)
jin.__weight -= 20

print(jin.__weight)
```

```python
class Person():
  def __init__(self, name, age, hobby, weight):
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.hobby = hobby
    self.__weight = weight # 변수 앞에 __ 를 붙이면 비공개 속성으로 변경

  def diet(self, loss_weight):
    self.__weight -= loss_weight
    print("{} kg 입니다.".format(self.__weight))

jin = Person('진', 22, '운동', 78)
jin.diet(20)
```

```python
class Person():
  def __Hello(self):
    print('hello')

  def hi(self):
    self.__Hello()

jin = Person()
jin.hi()
jin.__Hello() # 비공개 메서드라서 호출이 안된다.
```
