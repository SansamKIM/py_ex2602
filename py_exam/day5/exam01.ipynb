```python
# 클래스 속성
class Person():
    folder = [] # 클래스 속성 -> 클래스에 속한다.

    def add_files(self, file):
        Person.folder.append(file) #

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)

# ['txt', 'img']
# ['txt', 'img']
# -> 클래스 속성은 모든 인스턴스가 공유하는 속성 : 모든 인스턴스에서 접근 가능
```

```python
# 인스턴스 속성

class Person:
  def __init__(self):
    self.folder = [] # 인스턴스 속성

  def add_files(self, file):
    self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 비공개 클래스 속성

class Person:
  __weight = 78

  def print_weight(self):
    print(Person.__weight)

jin = Person()
jin.print_weight()

print(jin.__weight)
```
