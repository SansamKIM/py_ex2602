```python
# 파이썬 유니코드

# 1. 문자열은 컴퓨터 내부에서 숫자(유니코드)로 저장된다. => 문자를 숫자로 표현하는 방식
# 내부적으로는 각 글자를 숫자로 바꿔 저장한다. => 숫자 체계를 유니코드(Unicode)

text = "파이썬"

print(text)
print(len(text))
```

```python
# 2. 문자ㅡ 숫자 변환
print(ord('가'))
print(chr(44032))
```

```python
# 3. 문자열을 바이트(byte)로 변환
# encode : 문자열을 바이트(byte)로 바꾸는 작업

encoded = text.encode("utf-8")
print(encoded)
print(type(encoded))
```

```python
# 4. 바이트를 다시 문자열로 변환 (decode)
decoded = encoded.decode("utf-8")
print(decoded)
print(type(decoded))
# 파일을 읽을 때 인코딩 옵션(utf-8같은)이 중요하다
```
